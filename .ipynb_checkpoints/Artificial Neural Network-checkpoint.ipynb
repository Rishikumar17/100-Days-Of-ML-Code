```python
#Importing Libraries
#For Preprocessing the data
import pandas as pd
#For building models
import tensorflow as tf
#For visualization
import matplotlib.pyplot as plt
#For mathematical computations
import numpy as np
#For building models using tensorflow as backend
import keras
#Initializing the neural network
from keras.models import Sequential
#For making layers
from keras.layers import Dense
#Charts should be inline for jupyter notebook
%matplotlib inline
#Wrapping sklearn K-fold into keras
from keras.wrappers.scikit_learn import KerasClassifier
#Use for k-fold cross validation
from sklearn.model_selection import cross_val_score
#Scaling the dataset values to fall in a range
from sklearn.preprocessing import StandardScaler
#Splitting the dataset into train and test data
from sklearn.model_selection import train_test_split
#Dropout functions
from keras.layers import Dropout
#
from sklearn.model_selection import GridSearchCV
```

```python
#Reading the dataset
dataset = pd.read_csv("./Churn_Modelling.csv")
#Lets have a view of the dataset
dataset.head()
```

```python
#Getting Independant variables
X = dataset.iloc[:,3:13]
#Taking Dependant Variables
Y = dataset.iloc[:,13]
```

```python
#Getting dummy values for Geography
geography = pd.get_dummies(dataset["Geography"], drop_first=True)
#Getting dummy values for Gender
gender = pd.get_dummies(dataset["Gender"], drop_first=True)
#Adding the dummy columns with the existing independant variables
current_dataset = pd.concat([X, geography, gender], axis=1)
```

```python
#Removing all the categorical values since we have the dummy values
current_dataset.drop(["Gender","Geography"], axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(current_dataset,Y, test_size=0.2, random_state = 0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#Make an object of it
classifier = Sequential()
#Adding the input layer and the first hidden layer
#Note: The hidden layer count is more often not taken as (input layer node + output layer node)/2
classifier.add(Dense(output_dim = 6, kernel_initializer = 'uniform', activation='relu', input_dim = 11))
classifier.add(Dropout(rate = 0.1))
#Adding the second hidden layer
classifier.add(Dense(output_dim = 6, activation="relu", kernel_initializer= "uniform"))
classifier.add(Dropout(rate = 0.1))
#Adding the last output layer
classifier.add(Dense(output_dim = 1, kernel_initializer="uniform", activation="sigmoid"))
#Compiling the ANN
classifier.compile(optimizer="adam", metrics=["accuracy"], loss="binary_crossentropy")
#Fitting the ANN
classifier.fit(X_train,y_train,batch_size=10,epochs=100)
```

```python
#Getting the predictions
y_pred = classifier.predict(X_test)
#Changing the value of y_pred n boolean format of whether the customer is leaving or not leaving the bank
y_pred = (y_pred > 0.5)
#Getting the count of customer leaving and not leaving the bank
count = 0
for index,i in enumerate(y_pred):
    if y_pred[index] > 0.5:
        count = count + 1
print("Customer Not Leaving the Bank",count)
print("Customer Leaving the Bank",len(y_pred)-count)
```

```python
#Print the confusion matrix
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(kernel_initializer='uniform', activation='relu', input_dim = 11, output_dim = 6))
    classifier.add(Dense(kernel_initializer='uniform', activation='relu', output_dim = 6))
    classifier.add(Dense(output_dim=1, kernel_initializer='uniform', activation = 'sigmoid'))
    classifier.compile(optimizer="adam",metrics=["accuracy"],loss="binary_crossentropy")
    return classifier
```

```python
#Evaluating the ANN
classifier = KerasClassifier(build_fn=build_classifier,batch_size=10,epochs=100)
accuracies = cross_val_score(estimator=classifier, cv=10, X=X_train, n_jobs=-1, y=y_train)
mean = accuracies.mean()
variance = accuracies.std()
```

```python
#Tuning the ANN
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(kernel_initializer='uniform', activation='relu', input_dim = 11, output_dim = 6))
    classifier.add(Dense(kernel_initializer='uniform', activation='relu', output_dim = 6))
    classifier.add(Dense(output_dim=1, kernel_initializer='uniform', activation = 'sigmoid'))
    classifier.compile(optimizer=optimizer,metrics=["accuracy"],loss="binary_crossentropy")
    return classifier

classifier = KerasClassifier(build_fn=build_classifier)
parameters = {"batch_size":[25,32],
              "epochs":[100,500],
              "optimizer":["adam","rmsprop"]}
grid_search = GridSearchCV(estimator=classifier,param_grid=parameters,scoring='accuracy',cv=10)
grid_search = grid_search.fit(X_train, y_train)
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
```
