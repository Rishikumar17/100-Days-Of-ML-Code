```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
```

```python
dataset = pd.read_csv("./Mall_Customers.csv")
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,[3,4]].values
```

```python
dendogram = sch.dendrogram(sch.linkage(X,method = 'ward'))
plt.title('Dendograms')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters=5, affinity='euclidean',linkage='ward')
predictions = clustering.fit_predict(X)
```

```python
predictions
```

```python
#Visualisation of the clusters
plt.scatter(X[predictions == 0, 0], X[predictions == 0, 1], s=100, c='red', label = 'Careful')
plt.scatter(X[predictions == 1, 0], X[predictions == 1, 1], s=100, c='blue', label = 'Standard')
plt.scatter(X[predictions == 2, 0], X[predictions == 2, 1], s=100, c='green', label = 'Target')
plt.scatter(X[predictions == 3, 0], X[predictions == 3, 1], s=100, c='cyan', label = 'Careless')
plt.scatter(X[predictions == 4, 0], X[predictions == 4, 1], s=100, c='magenta', label = 'Sensible')
plt.xlabel("Annual Income")
plt.ylabel('Spending Power')
plt.legend()
plt.title('Plotting the elbow graph')
plt.show()
```
