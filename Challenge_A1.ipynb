```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("./Final_Version_College_data.csv", encoding="ISO-8859-1")
```

```python
data.head()
data = data.fillna(0)
```

```python
data.shape
```

```python
data.info()
```

```python
sumcs = {"2013":0,"2014":0,"2015":0,"2016":0,"2017":0}
sumnoncs = {"2013":0,"2014":0,"2015":0,"2016":0,"2017":0}
for i in range(len(data)):
    if "Computer Science" in str(data["major_degree"][i]):
        print(sumcs["2013"] + int(data["2013"][i]))
        sumcs["2013"] =  sumcs["2013"] + int(data["2013"][i])
        sumcs["2014"] =  sumcs["2014"] + int(data["2014"][i])
        sumcs["2015"] =  sumcs["2015"] + int(data["2015"][i])
        sumcs["2016"] =  sumcs["2016"] + int(data["2016"][i])
        sumcs["2017"] =  sumcs["2017"] + int(data["2017"][i])
    else:
        sumnoncs["2013"] =  sumnoncs["2013"] + int(data["2013"][i])
        sumnoncs["2014"] =  sumnoncs["2014"] + int(data["2014"][i])
        sumnoncs["2015"] =  sumnoncs["2015"] + int(data["2015"][i])
        sumnoncs["2016"] =  sumnoncs["2016"] + int(data["2016"][i])
        sumnoncs["2017"] =  sumnoncs["2017"] + int(data["2017"][i])
```

```python
sumcs
```

```python
sumnoncs
```

```python
plt.plot( sumcs.keys(),sumcs.values(), color='red')
plt.xlabel("Year")
plt.ylabel("Number of registrations")
plt.title("Trend for CS and Non-CS")
plt.plot( sumnoncs.keys(),sumnoncs.values(), color='green')
```
