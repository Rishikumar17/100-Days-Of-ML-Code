```python
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv("./Market_Basket_Optimisation.csv", header=None)
```

```python
dataset.head()
```

```python
dataset.isna().sum()
```

```python
sns.heatmap(dataset.isna(),yticklabels=False)
```

```python
transactions = []
```

```python
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
from apyori import apriori
```

```python
#Min_Support is calculated since we want atleast 3 purchases per day and so week its 21 divide by total number of observations
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.02, min_lift = 3, min_length=2)
```

```python
results = list(rules)
```

```python
for i in range(len(results)):
    for j in range(0,3):
        print(results[0])
```
